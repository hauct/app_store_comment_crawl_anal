```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome() 
url = 'https://play.google.com/store/apps/details?id=vng.games.thienlong.kiemhiep3d.mobile'
driver.get(url)

# Click the button with the specified class name
button = driver.find_element(By.CSS_SELECTOR, "button[class='VfPpkd-LgbsSe VfPpkd-LgbsSe-OWXEXe-dgl2Hf ksBjEc lKxP2d LQeN7 aLey0c']")
button.click()
time.sleep(2)

# Click button sort by
button_div = driver.find_element(By.CSS_SELECTOR, "div.D3Qfie.VfPpkd-ksKsZd-XxIAqe.KKjvXb.kofMvc.EFMXQ#sortBy_1")
button_div.click()
time.sleep(2)

# Click option Newest
# Find the span element with the specified class and aria-label
span_element = driver.find_element(By.CSS_SELECTOR, "span.z80M1.NmX0eb.KnEF3e[aria-label='Newest']")
span_element.click()
time.sleep(5)

# Function to scroll the target div until the end
def scroll_target_div():
    # Find the parent div with class name "odk6He"
    target_div = driver.find_element(By.CLASS_NAME, "fysCi")
    
    # Get initial scroll height
    last_height = driver.execute_script("return arguments[0].scrollHeight", target_div)
    
    scrolls = 0
    while True:
        # Scroll the target div to the bottom
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", target_div)
        
        # Wait for a short time for the page to load
        time.sleep(2)
        
        # Get updated scroll height
        new_height = driver.execute_script("return arguments[0].scrollHeight", target_div)
        
        # If scroll height hasn't changed, we've reached the bottom of the div
        if (new_height == last_height)|(scrolls == 200):
            break
        
        # Update last scroll height
        last_height = new_height
        scrolls = scrolls + 1

# Scroll down the specific div element
scroll_target_div()

# Get the page source after clicking the button
page_source = driver.page_source

# Parse the page source with Beautiful Soup
soup = BeautifulSoup(page_source, 'html.parser')

# Close the browser
driver.quit()
```

```python
from datetime import datetime

# get list review id
review_header = soup.find_all("header", class_= 'c1bOId')
review_ids = [x.get('data-review-id') for x in review_header]

# get list review stars
review_img = soup.find_all("div", class_= 'iXRFPc')
review_stars = [int(x.get('aria-label').split()[1]) for x in review_img]

# get list review date
review_span = soup.find_all("span", class_= 'bp9Aid')
review_dates = [datetime.strptime(x.text.strip(), '%B %d, %Y') for x in review_span]

# get list review content
review_text = soup.find_all("div", class_= 'h3YV2d')
review_contents = [x.text.strip() for x in review_text]

# get list review thumbs up
review_js = soup.find_all("div", {'jscontroller':'SWD8cc'})
review_thumbs = [int(x.get('data-original-thumbs-up-count')) for x in review_js]
```

```python
reviews_dict = []
for review_id, review_star, review_date, review_content, review_thumb in zip(review_ids, review_stars, review_dates, review_contents, review_thumbs):
    review_dict = {
        "review_id": review_id,
        "review_star": review_star,
        "review_date": review_date,
        "review_content": review_content,
        "review_thumb": review_thumb
    }
    reviews_dict.append(review_dict)

import pandas as pd
df = pd.DataFrame(reviews_dict[3:]) # remove 3 first preview reviews
```
